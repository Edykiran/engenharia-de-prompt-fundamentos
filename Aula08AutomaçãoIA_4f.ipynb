# Missão 1:

Criar um programa que organizae arquivos bagunçados em pastas automaticamente, basando-se apenas na extensão do arquivo (ex: .pdf, ,jpg) Utilize biblioteca os ou shuntil

### Organizador de Arquivos por Extensão

Este script Python automatiza a organização de arquivos em um diretório, movendo-os para subpastas que correspondem às suas extensões. Por exemplo, todos os arquivos `.pdf` serão movidos para uma pasta `PDF`, e arquivos `.jpg` para uma pasta `JPG`.

```python
import os
import shutil

def organize_files_by_extension(source_directory, destination_directory):
    """
    Organiza arquivos de um diretório de origem em subpastas no diretório de destino,
    baseando-se nas extensões dos arquivos.
    """
    # Verifica se o diretório de origem existe
    if not os.path.exists(source_directory):
        print(f"Erro: O diretório de origem '{source_directory}' não foi encontrado. Por favor, crie-o ou forneça um caminho válido.")
        return

    if not os.path.exists(destination_directory):
        os.makedirs(destination_directory)
        print(f"Diretório de destino '{destination_directory}' criado.")

    for filename in os.listdir(source_directory):
        source_path = os.path.join(source_directory, filename)

        # Ignora diretórios, processa apenas arquivos
        if os.path.isfile(source_path):
            # Obtém o nome do arquivo e a extensão
            _, extension = os.path.splitext(filename)
            extension = extension.lower() # Converte para minúsculas para padronização

            # Se não houver extensão (ex: 'arquivo_sem_extensao'), use 'NO_EXTENSION'
            if not extension:
                folder_name = 'NO_EXTENSION'
            else:
                # Remove o ponto da extensão para usar como nome da pasta
                folder_name = extension[1:].upper()

            destination_folder = os.path.join(destination_directory, folder_name)

            # Cria a pasta de destino se ela não existir
            if not os.path.exists(destination_folder):
                os.makedirs(destination_folder)
                print(f"Pasta '{destination_folder}' criada para arquivos {extension}.")

            destination_path = os.path.join(destination_folder, filename)

            try:
                shutil.move(source_path, destination_path)
                print(f"Movido '{filename}' para '{destination_folder}'")
            except Exception as e:
                print(f"Erro ao mover '{filename}': {e}")

# --- Uso do programa ---
# Defina o diretório de origem onde seus arquivos bagunçados estão
# Exemplo: Se você tem arquivos em '/content/unorganized_files', defina source_directory = '/content/unorganized_files'
source_directory = './unorganized_files' # Altere para o seu diretório de origem

# Defina o diretório de destino onde as pastas organizadas serão criadas
# Exemplo: Se você quer que as pastas sejam criadas em '/content/organized_files', defina destination_directory = '/content/organized_files'
destination_directory = './organized_files' # Altere para o seu diretório de destino


print(f"Iniciando organização de arquivos de '{source_directory}' para '{destination_directory}'...")
organize_files_by_extension(source_directory, destination_directory)
print("Organização concluída!")
```

#### Como usar:

1.  **Crie os diretórios:** Certifique-se de que o `source_directory` existe e contém os arquivos que você deseja organizar.
2.  **Modifique as variáveis:** Altere `source_directory` e `destination_directory` no código acima para os caminhos de suas pastas.
3.  **Execute a célula:** Clique no botão de execução para rodar o script.

# Missão 2

Desenvolver um script que consulte uma API pública (ViaCEP) e retorne as informações de endereço formatadas de maneira legível.

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço de um determinado CEP.
    Retorna um dicionário com os dados do endereço ou None em caso de erro.
    """
    # Remove quaisquer caracteres não numéricos do CEP
    cep = ''.join(filter(str.isdigit, cep))

    # Verifica se o CEP tem o tamanho correto (8 dígitos)
    if len(cep) != 8:
        print(f"Erro: O CEP '{cep}' é inválido. Ele deve conter 8 dígitos numéricos.")
        return None

    url = f"https://viacep.com.br/ws/{cep}/json/"

    try:
        response = requests.get(url)
        response.raise_for_status() # Lança um erro para status de resposta HTTP ruins (4xx ou 5xx)

        data = response.json()

        # Verifica se o CEP não foi encontrado (ViaCEP retorna 'erro': true)
        if 'erro' in data and data['erro']:
            print(f"Erro: CEP '{cep}' não encontrado.")
            return None

        return data

    except requests.exceptions.RequestException as e:
        print(f"Erro de conexão ao consultar o CEP '{cep}': {e}")
        return None
    except ValueError:
        print(f"Erro: A resposta da API para o CEP '{cep}' não é um JSON válido.")
        return None

def formatar_endereco(endereco_data):
    """
    Formata os dados de endereço em uma string legível.
    """
    if not endereco_data:
        return "Nenhuma informação de endereço disponível."

    formatted_string = (
        f"CEP: {endereco_data.get('cep', 'N/A')}\n"
        f"Logradouro: {endereco_data.get('logradouro', 'N/A')}\n"
        f"Complemento: {endereco_data.get('complemento', 'N/A')}\n"
        f"Bairro: {endereco_data.get('bairro', 'N/A')}\n"
        f"Localidade: {endereco_data.get('localidade', 'N/A')} - {endereco_data.get('uf', 'N/A')}\n"
        f"DDD: {endereco_data.get('ddd', 'N/A')}\n"
        f"IBGE: {endereco_data.get('ibge', 'N/A')}"
    )
    return formatted_string

# --- Exemplo de Uso ---
if __name__ == '__main__':
    # Solicita o CEP ao usuário
    cep_input = input("Digite o CEP para consulta (somente números): ")

    print(f"\nConsultando CEP: {cep_input}...")
    endereco = consultar_cep(cep_input)

    if endereco:
        print("\n--- Informações do Endereço ---")
        print(formatar_endereco(endereco))
    else:
        print("\nNão foi possível obter as informações do endereço.")

    print("\n--- Testando com CEP inválido/inexistente ---")
    cep_invalido = "99999999"
    print(f"Consultando CEP inválido: {cep_invalido}...")
    endereco_invalido = consultar_cep(cep_invalido)
    if not endereco_invalido:
        print("\nTeste de CEP inválido tratado com sucesso.")

    cep_curto = "123"
    print(f"Consultando CEP curto: {cep_curto}...")
    endereco_curto = consultar_cep(cep_curto)
    if not endereco_curto:
        print("\nTeste de CEP com formato inválido (curto) tratado com sucesso.")
```

# Missão 3

Criar um programa autônomo que dispare alertas(mensagens de alerta no console ou e-mail simulado) unicamente quando uma condição crítica for atingida.

```python
#Versão 1
import time
import random
from datetime import datetime

# --- Configurações ---
LIMITE_CRITICO = 38.0
INTERVALO_CHECK = 2  # segundos

def disparar_alerta_console(valor):
    print(f"\n[ALERTA CRÍTICO] - {datetime.now().strftime('%H:%M:%S')}")
    print(f">>> CONDIÇÃO ATINGIDA: {valor:.2f}°C ultrapassa o limite de {LIMITE_CRITICO}°C! <<<\n")

def enviar_email_simulado(valor):
    print(f"[E-MAIL ENVIADO] Para: suporte@sistema.com | Assunto: Falha Crítica Detectada ({valor:.2f}°C)")

def monitorar_sistema(ciclos=5):
    """
    Monitora o sistema por um número determinado de ciclos para demonstração.
    """
    print(f"Iniciando monitoramento autônomo... (Limite: {LIMITE_CRITICO}°C)")
    alerta_ativo = False

    for i in range(ciclos):
        # Simula leitura de sensor
        leitura = random.uniform(30.0, 42.0)
        status = "CRÍTICO" if leitura > LIMITE_CRITICO else "NORMAL"
        print(f"Ciclo {i+1}: {leitura:.2f}°C - Status: {status}")

        if leitura > LIMITE_CRITICO:
            if not alerta_ativo:
                disparar_alerta_console(leitura)
                enviar_email_simulado(leitura)
                alerta_ativo = True
        else:
            if alerta_ativo:
                print(">> Sistema estabilizado. Resetando alertas. <<")
                alerta_ativo = False

        time.sleep(INTERVALO_CHECK)

# Execução
if __name__ == '__main__':
    monitorar_sistema(ciclos=10)
```

```python
# Versão 2
import time
import random
from datetime import datetime

class MonitoramentoTermico:
    def __init__(self, limite=38.0, intervalo=2):
        self.limite_critico = limite
        self.intervalo_check = intervalo
        self.alerta_ativo = False

    def _obter_leitura(self):
        """Simula a leitura de um sensor real."""
        return random.uniform(30.0, 42.0)

    def _disparar_alertas(self, valor):
        """Centraliza o disparo de notificações."""
        hora = datetime.now().strftime('%H:%M:%S')
        print(f"\n[ALERTA CRÍTICO] - {hora}")
        print(f">>> CONDIÇÃO ATINGIDA: {valor:.2f}°C ultrapassa o limite de {self.limite_critico}°C! <<<")
        print(f"[E-MAIL ENVIADO] Para: suporte@sistema.com | Assunto: Falha Crítica ({valor:.2f}°C)\n")

    def executar(self, ciclos=10):
        print(f"Iniciando monitoramento... (Limite: {self.limite_critico}°C)")

        for i in range(1, ciclos + 1):
            leitura = self._obter_leitura()
            status = "CRÍTICO" if leitura > self.limite_critico else "NORMAL"
            print(f"Ciclo {i}: {leitura:.2f}°C - Status: {status}")

            if leitura > self.limite_critico:
                if not self.alerta_ativo:
                    self._disparar_alertas(leitura)
                    self.alerta_ativo = True
            else:
                if self.alerta_ativo:
                    print(">> Sistema estabilizado. Resetando alertas. <<")
                    self.alerta_ativo = False

            time.sleep(self.intervalo_check)

if __name__ == '__main__':
    # Instancia e executa o monitor
    monitor = MonitoramentoTermico(limite=38.0, intervalo=1)
    monitor.executar(ciclos=5)
```
